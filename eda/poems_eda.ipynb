```python
import pandas as pd
```

```python
path = "../data/PoetryFoundationData.csv"
df = pd.read_csv(path)
print("Loaded {} poems.".format(len(df)))
```

```python
# select and rename columns
df = df[['Title', 'Poem', 'Poet', 'Tags']]
df.columns = df.columns.str.lower()
```

```python
# print a few poems
for i in range(15):
    print(df['title'][i])
    print(df['poem'][i])
    print("By {}.".format(df['poet'][i]))
    print('=' * 20)
    
```

```python
# clean poem string
df['poem_clean'] = df['poem'].str.replace('\r', '')
df['poem_clean'] = df['poem_clean'].str.replace('\n', ' ')

# remove whitespace
df['poem_clean'] = df['poem_clean'].str.strip()
df = df[df['poem_clean'] != ""]
```

```python
# get word count
df['poem_len'] = df['poem_clean'].str.split(' ').apply(len)
print("Poems length stats")
print(df['poem_len'].describe())

# keep poems that are less or equal to 500 words
df = df[df['poem_len'] <= 500]
```
